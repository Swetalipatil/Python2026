6. Write a Python function that accepts a string and counts the number of upper and lower case letters.
	Sample String : 'The quick Brow Fox'
	Expected Output :
	No. of Upper case characters : 3
	No. of Lower case Characters : 12



```python
def count_upper_lower(s):
    upper = lower = 0
    for c in s:
        if c.isupper():
            upper += 1
        elif c.islower():
            lower += 1
    return upper, lower
sample_string = 'The quick Brow Fox'
upper, lower = count_upper_lower(sample_string)
print("No. of Upper case characters:", upper)
print("No. of Lower case characters:", lower)
```

7. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

	Sample List : [1,2,3,3,3,3,4,5]
	Unique List : [1, 2, 3, 4, 5]


```python
def unique_list_ordered(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique
sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
print(unique_list_ordered(sample_list))
```

8. Write a Python function that checks whether a passed string is a palindrome or not.

	Note: A palindrome is a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.



```python
def is_palindrome(s):
    s = s.replace(" ", "").lower()
    return s == s[::-1]

print(is_palindrome("madam"))        # True
print(is_palindrome("nurses run"))   # True
print(is_palindrome("hello"))        # false
```

9. Write a Python program that accepts a hyphen-separated sequence of words as input and prints the words in a hyphen-separated sequence after sorting them alphabetically.

	Sample Items : green-red-yellow-black-white
	Expected Result : black-green-red-white-yellow



```python
def sort_words(text):
    words = text.split('-')
    return '-'.join(sorted(words))

# Input
user_input = input("Enter words: ")
print(sort_words(user_input))
```

13. Write a Python program to sort a list of tuples using Lambda.

	Original list of tuples:
	[('English', 88), ('Science', 90), ('Maths', 97), ('Social sciences', 82)]
	Sorting the List of Tuples:
	[('Social sciences', 82), ('English', 88), ('Science', 90), ('Maths', 97)]


```python

subjects = [('English', 88), ('Science', 90), ('Maths', 97), ('Social sciences', 82)]

sorted_subjects = sorted(subjects, key=lambda x: x[1])
print("Sorting the List of Tuples:")
print(sorted_subjects)
```

19. Write a Python program to rearrange positive and negative numbers in a given array using Lambda.

	Original arrays:
	[-1, 2, -3, 5, 7, 8, 9, -10]
	Rearrange positive and negative numbers of the said array:
	[2, 5, 7, 8, 9, -10, -3, -1]



```python
arr = [-1, 2, -3, 5, 7, 8, 9, -10]
positives = [x for x in arr if x > 0]
negatives = [x for x in arr if x < 0][::-1]
rearranged = positives + negatives
print(rearranged)
```

21. Write a Python program to find numbers divisible by nineteen or thirteen from a list of numbers using Lambda.

	Orginal list:
	[19, 65, 57, 39, 152, 639, 121, 44, 90, 190]
	Numbers of the above list divisible by nineteen or thirteen:
	[19, 65, 57, 39, 152, 190]


```python

numbers = [19, 65, 57, 39, 152, 639, 121, 44, 90, 190]
mapped = map(lambda x: x if x % 13 == 0 or x % 19 == 0 else None, numbers)
result = [x for x in mapped if x is not None]
print("Original list:", numbers)
print("Numbers divisible by 13 or 19:", result)
```
